## Imagens Coloridas

Como já vimos, as imagens manipuladas pelo `scikit-image` são simplesmente arrays do Numpy. Consequentemente, uma grande parte das operações sobre as imagens consistem do uso das funcionalidades do Numpy.

As imagens coloridas seguem este mesmo conceito: são arrays do Numpy, mas com uma dimensão adicional para os canais:

```python
%matplotlib inline
```

```python
from skimage import data
cat = data.chelsea()
type(cat)
```

```python
import numpy as np
```

```python
cat.shape
```

```python
cat[10,20]
```

```python
from pylab import rcParams

img = data.astronaut()

rcParams['figure.figsize'] = 10, 10
plt.xticks([]), plt.yticks([])
plt.imshow(img),plt.title("Imagem original")
img.shape
```

```python
from skimage.exposure import equalize_hist, rescale_intensity

rcParams['figure.figsize'] = 20, 20

plt.subplot(131),plt.imshow(equalize_hist(img)),plt.title("Imagem com Equalização"),plt.xticks([]), plt.yticks([])

plt.subplot(132),plt.imshow(rescale_intensity(img)),plt.title("Imagem com Normalização"),plt.xticks([]), plt.yticks([])
```

```python
from skimage import filters
from skimage.color.adapt_rgb import adapt_rgb, each_channel, hsv_value
from skimage.filters import gaussian

#Gerando imagem aguçada 
@adapt_rgb(each_channel)
def sobel_each(image):
    return filters.sobel(image)

img_aguçada = sobel_each(img)

#Gerando imagem suavizada
sigma = 2
img_suavizada = gaussian(img,sigma=3,multichannel=True)

#Plot
rcParams['figure.figsize'] = 20, 20

plt.subplot(131),plt.imshow(img_aguçada),plt.title('Imagem aguçada com filtro sobel'),plt.xticks([]), plt.yticks([])

plt.subplot(132),plt.imshow(img_suavizada),plt.title('Imagem suavizada com filtro gausian : %s' % (sigma)),plt.xticks([]), plt.yticks([])
```
